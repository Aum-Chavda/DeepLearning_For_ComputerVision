# Comparison of all Regularization 

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

The following code will load all `*.pickle` files that you created in the other notebooks, and plot the saved losses.

```python
import pickle
import glob

# Get the default colors of matplotlib
prop_cycle = plt.rcParams['axes.prop_cycle']
colors = prop_cycle.by_key()['color']

plt.figure(figsize=(10, 10))
for f, color in zip(sorted(glob.glob('*.pickle')), colors):
    with open(f, 'rb') as handle:
        train_losses, val_losses = pickle.load(handle)
    label = " ".join(f.split('.')[0].split('_')[1:])
    plt.plot(train_losses, '--', label=f"{label} (train)", c=color)
    plt.plot(val_losses, label=f"{label} (val)", c=color)

# plt.yscale('log')
plt.legend()
plt.show()
```
